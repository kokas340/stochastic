```python
# H_0 = M_0=1
# H_1 = M_0!=1
#N~(M_0,standard deviation^2) for i = 1,2,3,4,5,..,9 
#-1.7, 2.2, 3.2, -0.8, 2.0, 0.7, -2.2, 0.5, 3.0 
# the test statistic is the sample mean
sample = np.array([-1.7, 2.2, 3.2, -0.8, 2.0, 0.7, -2.2, 0.5, 3.0])
# Null hypothesis value
mu0 = 1  # Null hypothesis: the mean is 1
# calculate sample statistics
n = len(sample)
xbar = np.mean(sample)
s = np.std(sample, ddof=1)
# calculate test statistic
z = (xbar - mu0) / (s / np.sqrt(n))
print("z = ", z)


#Significance level
alpha = 0.1 

#Degrees of freedom
df = len(sample) - 1

#Calculate the critical value (two-tailed test)
cv = t.ppf(1.0 - alpha / 2, df)
print("cv = ", cv)
```
