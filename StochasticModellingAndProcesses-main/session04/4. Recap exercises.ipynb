```python
import sympy as sp

x, c, u = sp.symbols('x c u')

sp.simplify(sp.integrate(c*u**4, (u, -1, x)))
```

```python
def f(u):
    return c*u**4
```

```python
sp.simplify(sp.integrate(f(u), (u, -1, x)))
```

```python
sp.solve(sp.integrate(c*x**4, (x, -1, 1)) -1, c)[0]
```

```python
sp.solve(sp.integrate(f(u), (u, -1, 1)) -1, c)[0]
```
