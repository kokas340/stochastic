```python
# here we have a binomail distribution with (n = from 0 - 1000, p = 0.01)
# we are looking for the n that makes the following equation true 
# binome.sf(0, n, 0.01) >= 0.9 
# we are looking for n sample size where this equation is true

for n in range(0 , 1000):
    result = binom.sf(0, n, 0.01)
    if result >= 0.9:
        print(n)
        break
```
